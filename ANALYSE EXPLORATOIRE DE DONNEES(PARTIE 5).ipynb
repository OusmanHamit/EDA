ANALYSE EXPLORATOIRE DE DONNEES/ LES JOINTURES AVEC PANDAS(PARTIE 5)
- INNNER JOIN 
- LEFT JOIN
- RIGHT JOIN
- FULL OUTER JOIN

```python
import pandas as pd 

df1 = pd.DataFrame({'PAYS':['Tchad','Libye','Soudan','Algerie','Tunisie','Chine'],
                    'CAPITALE':['Ndjamena','Tripoli','Khartoum','Alger','Tunis','Pékin'],
                    'POPULATION':['16M','6M','43M','43M','11M','1,402 milliard'],
                    'LANGUES':['Arabe/francais','Arabe','Arabe','Arabe','Arabe','Chinois'],
                    'CONTINENT':['Afrique','Afrique','Afrique','Afrique','Afrique','Asie'],
                   })

df2 = pd.DataFrame({'PAYS':['Tunisie','Tchad','Soudan','Mali','Senegal'],
                    'SUPERFICIE':['163 610 km²',
                                  '1,284 million km²',
                                  '1,886 million km²',
                                  '1,24 million km²',
                                  '196 722 km²'
                                ],
                    'DEVISE':['Dinar tunisien',
                              'Franc CFA',
                              'Livre soudanaise',
                              'Franc CFA',
                              'Franc CFA'
                            ]
                   })
```

```python
df1
```

```python
df2
```

-----------------------------

1.  INNER JOIN OU INTERSECTION

 - Au moins une colonne commune ayant les memes valeurs dans les deux DataFrames

```python
inner = pd.merge(df1, df2, on = 'PAYS', how = 'inner')
inner 
```

-----------------------------

2. LEFT JOIN avec:

    - Pas de colonnes ayant le meme labels
    - Avec quelques valeurs communes

```python
df3 = pd.DataFrame({'COUNTRIES':['Tunisie','Tchad','Soudan','Mali','Senegal'],
                    'SUPERFICIE':['163 610 km²',
                                  '1,284 million km²',
                                  '1,886 million km²',
                                  '1,24 million km²',
                                  '196 722 km²'
                                ],
                    'DEVISE':['Dinar tunisien',
                              'Franc CFA',
                              'Livre soudanaise',
                              'Franc CFA',
                              'Franc CFA'
                            ]
                   })
df3
                   
```

```python
left = pd.merge(df1, df3, how = 'left', left_on = 'PAYS', right_on = 'COUNTRIES', indicator = True)
left 
```

-----------------------------

3. RIGHT JOIN avec:

    - Ayant au moins une colonne en commun
    - Quelques valeurs communes pour ces colonnes

```python
right = pd.merge(df1, df2, how = 'right', on = 'PAYS', indicator = 'PROVENANCE')
right
```

-----------------------------

4. OUTER JOIN / FULL OUTER JOIN avec:
- au moins une colonne en partage(PAYS) ayant les memes valeurs dans les deux datasets

```python
outer = pd.merge(df1, df2, how = 'outer', on = 'PAYS', sort = True)
outer
```
