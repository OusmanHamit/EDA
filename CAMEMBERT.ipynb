```python
from matplotlib import pyplot as plt
```

- **DONNEES A VISUALISER** 

```python
categories = ['Celibataire(s)', 'Marie(es)', 'Divorce(es)', 'Veuf(ves)', 'Autres']
frequences = [200, 100, 97, 50, 150]
```

```python
plt.figure(figsize = (8, 8))
plt.style.use('dark_background')
plt.pie(frequences);
```

- **PERSONNALISER LE GRAPHIQUE** 

```python
colors = ['indigo', 'navy', 'red', 'gray', 'brown']
explode = [0.09, 0, 0, 0, 0]
```

```python
plt.figure(figsize = (8, 8))
plt.pie(frequences,
        labels = categories,
        colors = colors,
        explode = explode,
        autopct = '%1.2f%%')
plt.title('Diagramme en secteur')
plt.tight_layout()
plt.legend(loc = 2, title = 'Categories')
plt.savefig('camembert.png')
plt.show()
```
