```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set()
```

---

```python

sns.get_dataset_names()
```

```python
df = sns.load_dataset("taxis")
df.head()
```

---

- SEABORN HEATMAP

```python
plt.figure(figsize = (8,4))
sns.heatmap(df.select_dtypes(['number']).corr(), 
            annot = True, 
            cmap = 'coolwarm', 
            annot_kws = {'fontsize': 10}, 
            center = 0,
            vmin = -1, 
            lw = .4,
            vmax = 1, 
            cbar = True,)
plt.title("Heatmap: taxis dataset")
plt.savefig("heatmap.png")
plt.show()
```

---

- SEABORN CATPLOT

```python
df2 = sns.load_dataset("tips")
df2.head()
```

```python
sns.catplot(x = 'day',
            y = 'total_bill',
            hue = 'sex',
            data = df2,
            kind = 'box')
plt.savefig("boxplot.jpg")
plt.show()
```

```python
sns.catplot(x = 'day',
            y = 'total_bill',
            hue = 'sex',
            data = df2,
            kind = 'strip',
            palette = 'cool')
plt.savefig("strip.jpg")
plt.show()
```
